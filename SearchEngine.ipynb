## Web Crawler

```python

import requests
from bs4 import BeautifulSoup
import pandas as pd

def crawler():
    links = []
    data = {'publication':[],'title':[],'authors':[],'profilelink':[[]],'date':[]}
    for i in range(0,13):
        req1 = requests.get(f'https://pureportal.coventry.ac.uk/en/organisations/school-of-economics-finance-and-accounting/publications/?page={i}')
        soup1 = BeautifulSoup(req1.content,'lxml') 
        publicationlist = soup1.find_all('h3',class_='title')

        #Collecting Dates
        datelist = soup1.find_all('div',class_='result-container')
        for item in datelist:
          for x in item.find_all('span',class_='date'):
            data['date'].append(x.text)
        
        for item in publicationlist:
            # Collecting Publication Links
            for link in item.find_all('a',href=True,class_="link"):
                data['publication'].append(link['href'])

                #Opening publication page
                req2 = requests.get(link['href'])
                soup2 = BeautifulSoup(req2.content,'lxml')
                              
                # Collecting title
                title = soup2.find_all('h1')
                for item in title:
                   data['title'].append(item.text)

                # Collecting Authors names
                relatelist = soup2.find_all('p',class_="relations persons")
                for item in relatelist:
                    data['authors'].append(item.text) 

                    # Collecting Profile Links - Few more steps to manage more than one link with a single publication

                    listoflinks = soup2.find_all('a',href = True, class_="link person")
                    if(len(listoflinks)==0):
                        links=[] 
                        links.append("No Links")
                    for i in range(0,len(listoflinks)): 
                        links = [] 
                        for link in listoflinks:
                          links.append(link['href'])   
                    data['profilelink'].append(links) 
    
    del data['profilelink'][0]
    return(data)



corpus = crawler()
```

```python
import pandas as pd

df = pd.DataFrame(corpus)
```

Tokenization and Stemming

```python
# this function returns a list of tokenized and stemmed words of any text
def get_tokenized_list(doc_text):
    tokens = nltk.word_tokenize(doc_text)
    return tokens

# This function will performing stemming on tokenized words
def word_stemmer(token_list):
  ps = nltk.stem.PorterStemmer()
  stemmed = []
  for words in token_list:
    stemmed.append(ps.stem(words))
  return stemmed
```

Removing Stopwords

```python
# Function to remove stopwords from tokenized word list
def remove_stopwords(doc_text):
  cleaned_text = []
  for words in doc_text:
    if words not in stop_words:
      cleaned_text.append(words)
  return cleaned_text
```

Checking for single document

```python
#Check for single document
print(f"Orginal Document :{df['title'][0]}\n")
tokens = get_tokenized_list(df['title'][0])
print("WORD TOKENS:")
print(tokens)
doc_text = remove_stopwords(tokens)
print("\nAFTER REMOVING STOPWORDS:")
print(doc_text)
print("\nAFTER PERFORMING THE WORD STEMMING::")
doc_text = word_stemmer(doc_text)
print(doc_text)
```

Applying the same steps for all the documents

```python
cleaned_corpus = []
for doc in df['title']:
    tokens = get_tokenized_list(doc)
    doc_text = remove_stopwords(tokens)
    doc_text  = word_stemmer(doc_text)
    doc_text = ' '.join(doc_text)
    cleaned_corpus.append(doc_text)
print(cleaned_corpus)
```

```python
vectorizerX = TfidfVectorizer()
vectorizerX.fit(cleaned_corpus)
doc_vector = vectorizerX.transform(cleaned_corpus)
print(vectorizerX.get_feature_names_out())
```

```python
print(doc_vector.shape)
```

```python
newdf = pd.DataFrame(doc_vector.toarray(), columns=vectorizerX.get_feature_names_out())
print(newdf)
```

```python
query = str(input("Enter query :"))
new_query = query
query = get_tokenized_list(query)
query = remove_stopwords(query)
q = []
for w in word_stemmer(query):
  q.append(w)
q = ' '.join(q)
q
query_vector = vectorizerX.transform([q])
```

```python
# calculate cosine similarities
from sklearn.metrics.pairwise import cosine_similarity
cosineSimilarities = cosine_similarity(doc_vector,query_vector).flatten()
```

```python
related_docs_indices = cosineSimilarities.argsort()[:-6:-1]
print(f"Search Results for {new_query}\n")
    
for i in related_docs_indices:
  print(f"\nTittle : {df['title'][i]}")
  print(f"Publication Link : {df['publication'][i]}")
  print(f"Authors : {df['authors'][i]}")
  print(f"Date : {df['date'][i]}")
  print(f"Profile Link : {df['profilelink'][i]}\n")
```
